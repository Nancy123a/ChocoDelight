## Data Preprocessing

### Numeric String Preprocessing and Conversion Function

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

def preprocess_numeric_value(value):
    if isinstance(value, str):
        # Remove dots used as thousand separators and 
        # replace commas with dots
        value = value.replace('.', '').replace(',', '.')
        # Check if the resulting string is empty or non-numeric
        if value == '' or not value.replace('.', '', 1).replace('-', '', 1).isdigit():
            return np.nan
        return float(value)
    return value

# Select columns from 3rd to one before the last column
numeric_columns = random_production.columns[2:-1]  
## Numeric String Preprocessing and Conversion Function
random_production[numeric_columns] = random_production[numeric_columns].apply(lambda x: x.map(preprocess_numeric_value))
```

## Handling Outliers

### Function to identify outliers:

```python
def get_outlier_indices(column):
    # Calculate the first quartile (Q1), third quartile (Q3),
    # and interquartile range (IQR)
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    
    # Calculate the lower and upper bounds for outliers
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers based on the bounds
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers.index
```

### Number of outliers before handling

```python
def print_number_of_outliers(df):
    outliers_count = {col: len(get_outlier_indices(df[col])) for col in numeric_columns}
    
    for col, count in outliers_count.items():
        if count > 0:
            print(f"Number of outliers in {col}: {count}")
```

### IQR Winsorization - Skewed distribution (Cocoa_Roasting_Level):

```python
def modify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define limits based on IQR
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Apply winsorization manually using .loc for assignment
    df.loc[df[column] < lower_limit, column] = lower_limit
    df.loc[df[column] > upper_limit, column] = upper_limit
    
    return df[column]
```

```python
columns_to_modify = ['Cocoa_Roasting_Level']

for col in columns_to_modify:
    random_production[col] = modify_outliers(random_production, col)
```

### Z-score Winsorization:

```python
def modify_outliers_z(df, column):
    # Calculate mean and standard deviation
    mean = df[column].mean()
    std_dev = df[column].std()
    
    # Define threshold as Â±3 standard deviations
    lower_limit = mean - 2 * std_dev
    upper_limit = mean + 2 * std_dev
    
    # Apply winsorization
    df[column] = np.where(df[column] < lower_limit, lower_limit,
                          np.where(df[column] > upper_limit, upper_limit,
                                   df[column]))
    
    return df[column]
```

```python
columns_to_modify = ['Milk_Solid_Percentage',
                     'Barcode_Scan_Success',
                     'Label_Readability',
                     'Nutritional_Info_Accuracy', ]

for col in columns_to_modify:
    random_production[col] = modify_outliers_z(random_production, col)
```

## Feature Importance Analysis:
### Ascertain which features most influence a station's "not okay" status.

**PCA (Principal Component Analysis):**

PCA is a statistical technique used to reduce the dimensionality of a dataset while retaining as much variance (information) as possible. It transforms a set of correlated variables into a smaller set of uncorrelated variables called principal components.

**Cumulative Explained Variance:**

Cumulative explained variance in the context of PCA refers to the cumulative amount of variance explained by the principal components in descending order of their eigenvalues. It helps in understanding how much of the original dataset's variance is retained as the number of principal components increases.

**Cumulative Explained Variance Graph:**

This graph visually represents the cumulative explained variance against the number of principal components. Typically, it shows how much variance is explained as you add each successive principal component. It helps in deciding how many principal components to retain based on how much variance one wants to capture from the original dataset.

'explained_variance_ratio_' tells you the variance explained by each individual principal component.

'cumulative_variance_ratio' tells you the accumulated variance explained by adding each successive principal component to the previous ones.

```python
from sklearn.decomposition import PCA

# Separate data by 'Status' (1 for Okay, 0 for Not Okay)
okay_data = random_production[random_production['Status'] == 1]
not_okay_data = random_production[random_production['Status'] == 0]

# Perform PCA on the numeric columns of the combined dataset
pca = PCA()
pca.fit(random_production[numeric_columns])

# Get the explained variance ratio for each principal component
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio * 100, marker='o', linestyle='--', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('Cumulative Explained Variance by Principal Component')
plt.axhline(y=95, color='r', linestyle='-')  # Horizontal line at 95% cumulative variance
plt.axvline(x=np.argmax(cumulative_variance_ratio >= 0.80) + 1, color='r', linestyle='--')  # Vertical line at 80% explained variance
plt.grid(True)
plt.show()

# Find the number of components that explain at least 80% of the variance
optimal_components = np.argmax(cumulative_variance_ratio >= 0.80) + 1
print(f'Optimal number of principal components to retain 80% of the variance: {optimal_components}')
```

```python
# Analyze PCA components and feature loadings
component_names = [f"PC{i+1}" for i in range(len(explained_variance_ratio))]
pc_df = pd.DataFrame(pca.components_, columns=numeric_columns, index=component_names)
pc_df.T
```

## Use techniques such as recursive feature elimination or permutation importance.

### Recursive Feature Elimination (RFE)
Recursive Feature Elimination (RFE) is a technique that recursively removes features and builds a model on the remaining features. It ranks features by their importance, typically using the model's coefficient values. 

- 'estimator': Specifies the base estimator from which the model is built, in this case, logreg (logistic regression).
- 'n_features_to_select': Specifies the number of features to select, set to 1 here, meaning RFE will select the single most important feature.
- 'step': Controls the number of features to remove at each iteration. In this case, step=1 means one feature is eliminated in each iteration.

- RFE will iterate approximately n_features - 1 times because it starts with all features and eliminates one feature in each iteration until only one feature remains (the one selected as most important).

```python
from sklearn.feature_selection import RFE

# Initialize logistic regression model
logreg = LogisticRegression()

# Perform RFE with logistic regression
rfe = RFE(estimator=logreg, n_features_to_select=1, step=1)
rfe.fit(X_train, y_train)

# Print ranking of features
print("Ranking of features by RFE:")
for i, rank in enumerate(rfe.ranking_):
    print(f"Rank {rank}: {random_production[numeric_columns].columns[i]}")
```

- RFE ranks the features based on their importance. Lower rank numbers indicate higher importance.
- Nuts Inclusion rate, Shelf life test, sweetness level, and seal integrity have the most significant impact on predicting the "not okay" status.

**Logistic regression** is a statistical model used to predict binary outcomes based on predictor variables. It models the probability of the default class (typically labeled as 1) using the logistic function. The formula for logistic regression is:

Logistic regression is a statistical model used to predict binary outcomes based on predictor variables. It models the probability of the default class (typically labeled as 1) using the logistic function. The formula for logistic regression is:

$p(y=1 | \mathbf{x}) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n)}}$

- $p(y=1 | \mathbf{x})$: Probability of the event \( y=1 \) given predictors \( \mathbf{x} \).
- $\beta_0, \beta_1, \beta_2, \dots, \beta_n$: Coefficients estimated by the model.
- $x_1, x_2, \dots, x_n$: Predictor variables.
- $e$: Base of the natural logarithm (Euler's number).

In logistic regression, the log odds of the outcome (logit) is modeled as a linear combination of the predictors. The logistic function (sigmoid function) transforms this linear combination into a probability score between 0 and 1, which is then used to classify observations into one of two categories based on a threshold (often 0.5).

Sure, here are the definitions of each metric briefly with their formulas:

1. **Accuracy:**
   - Measures the proportion of correctly classified instances out of the total instances.
   - Formula: $\text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}$
     - $\text{TN}$: True Negatives (correctly predicted negative instances).
   - Accuracy provides an overall measure of the model's correctness in predicting all classes

2. **Precision:**
   - Measures the accuracy of positive predictions made by the model. - model is predicting positive
   - Formula: $\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}$
     - $\text{TP}$: True Positives (correctly predicted positive instances).
     - $\text{FP}$: False Positives (incorrectly predicted as positive, actually negative).
   - A high precision indicates that when the model predicts a class as positive, it is usually correct.

3. **Recall:**
   - Measures the ability of the model to identify all relevant instances of the positive class. - they are actually positive
   - Formula: $\text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}$
     - $\text{FN}$: False Negatives (incorrectly predicted as negative, actually positive).
   - A high recall indicates that the model is good at capturing instances of the class.

4. **F1 Score:**
   - Harmonic mean of precision and recall, providing a balance between the two.
   - Formula: $\text{F1 Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$
   - F1-score reaches its best value at 1 (perfect precision and recall) and worst at 0.
     
5. **Support:**
   - The number of actual occurrences of each class in the dataset.
   -  It provides insights into the distribution of instances across different classes. Classes with higher support have more instances in the dataset.


These metrics are essential in evaluating the performance of binary classification models, providing insights into their predictive power and reliability.

- **Macro Average**: Gives equal weight to each class, regardless of class size. It's useful when all classes are equally important.
- **Micro Average**: Considers the overall performance across all classes, weighted by the number of instances in each class. It's useful when class imbalance is present.


### Train logistic model

- class_weight='balanced' is a parameter that automatically adjusts the weights of classes inversely proportional to their frequencies. This means the model assigns higher weights to minority class instances and lower weights to majority class instances during training (reduce being biased towards the majority class). 

```python
def train_logistic_model(X_train, X_test, y_train, y_test):
    # Initialize model with balanced class weights
    model = LogisticRegression(class_weight='balanced', random_state=42)
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Plot classification results
    plot_classification_results(y_test, y_pred)
```

```python
# Function to plot classification results
def plot_classification_results(y_true, y_pred):
    # Generate classification report
    report = classification_report(y_true, y_pred, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    print('classification report')
    print(report)

    # Generate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    print('confusion matrix')
    print(cm)
    
    # Create figure with two subplots
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))

    # Plot classification report
    sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='Blues', fmt='.2f', ax=axes[0], annot_kws={"size": 16}, cbar=False)
    axes[0].set_title('Classification Report', fontsize=20)
    axes[0].set_yticklabels(axes[0].get_yticklabels(), rotation=0, fontsize=14)
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, fontsize=14)

    # Plot confusion matrix
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, 
                xticklabels=['Predicted Not Okay', 'Predicted Okay'],
                yticklabels=['Actual Not Okay', 'Actual Okay'], 
                ax=axes[1], annot_kws={"size": 16})
    
    axes[1].set_xlabel('Predicted', fontsize=18)
    axes[1].set_ylabel('Actual', fontsize=18)
    axes[1].set_title('Confusion Matrix', fontsize=20)
    axes[1].set_yticklabels(axes[1].get_yticklabels(), rotation=0, fontsize=14)
    axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, fontsize=14)

    plt.tight_layout()
    plt.show()
```

### Optuna (Bayesian optimization) with oversampling

Optuna optimizes hyperparameters using Bayesian methods and, with oversampling like SMOTE, improves model performance by tuning hyperparameters for better accuracy or F1-score, crucial for imbalanced datasets.

**Why Optuna:**
1 - Bayesian Optimization (probabilistic model of the objective function).
2- Efficiency:  because it iteratively selects hyperparameters based on the expected improvement criterion.
3- Automatic Logging and Visualization: It provides automatic logging of hyperparameter search trials and visualization tools to track and analyze the optimization process. 

- 'C' is inverse of regularization strength (smaller values specify stronger regularization)
- 'solver' is a parameter in logistic regression that determines the optimization algorithm. 'lbfgs' is a good choice for smaller datasets. 'liblinear' is suitable for handling datasets with a large number of samples or features.

```python
import optuna
import logging

optuna.logging.set_verbosity(optuna.logging.WARNING)
logging.basicConfig(level=logging.WARNING)

# Define your objective function
def objective(trial, X_train, X_test, y_train, y_test):
    # Define hyperparameters to tune
    params = {
        'C': trial.suggest_float('C', 1e-4, 1e2, log=True), 
        'solver': trial.suggest_categorical('solver', ['lbfgs', 'liblinear'])
    }

    # Create the logistic regression model
    model = LogisticRegression(class_weight='balanced', random_state=42, **params)
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Evaluate on the test set
    y_pred = model.predict(X_test)
    
    # Calculate F1 score
    f1 = f1_score(y_test, y_pred, average='binary', pos_label=0) ##optimize according to f1 score of minority class (maximize)
    
    return f1

# Example usage:
# Assume X and y are already defined and prepared
# Create an Optuna study object
study = optuna.create_study(direction='maximize')

# Pass X_train, X_test, y_train, and y_test to the objective function using a lambda function
study.optimize(lambda trial: objective(trial, X_train_res, X_test, y_train_res, y_test), n_trials=100)

# Print the best hyperparameters and best accuracy found
best_params = study.best_params
best_f1score = study.best_value
print(f"Best hyperparameters: {best_params}")
print(f"Best f1 score: {best_f1score}")
```

```python
best_model = LogisticRegression(class_weight='balanced', random_state=42, **best_params)
## fit on the balanced dataset
best_model.fit(X_train_res, y_train_res)

# Evaluate on the test set
y_pred = best_model.predict(X_test)
    
# Plot classification results
plot_classification_results(y_test, y_pred)
```

- Therefore, the model with both oversampling and Optuna optimization is preferable for identifying "not okay" cases (class 0), as it demonstrates better performance metrics aligned with the primary objective.

### Oversamling + Cross validation (4 folds)

- **K-fold Cross-Validation**: It involves splitting the dataset into K folds (or subsets) of approximately equal size.
- **Training and Validation**: The model is trained using K-1 folds and validated on the remaining fold. This process is repeated K times, with each fold used once as the validation data.
- **Performance Evaluation**: After K iterations, performance metrics such as accuracy, precision, recall, or F1-score are averaged across all folds to obtain a final performance estimate.

- **Benefits of K-fold cross validation on unbalanced data:** Reduce bias, better generalization, robust evaluation, and hyperparameters tuning.

```python
y_train_res = y_train_res.reset_index(drop=True)  # Reset index to avoid issues
```

```python
from sklearn.model_selection import StratifiedKFold
def kfold_cv(number_of_folds):

    # Define the classifier (e.g., Logistic Regression)
    clf = LogisticRegression(class_weight='balanced', **best_params, random_state=42)
    
    kf = StratifiedKFold(n_splits=number_of_folds, shuffle=True, random_state=42)
    
    # Lists to store results
    f1_scores = []
    cm_matrices = []  # To store confusion matrices
    classification_reports = []
    
    # Perform cross-validation
    for train_index, val_index in kf.split(X_train_res, y_train_res):
        X_train_fold, X_val_fold = X_train_res.iloc[train_index], X_train_res.iloc[val_index]
        y_train_fold, y_val_fold = y_train_res.iloc[train_index], y_train_res.iloc[val_index]
    
        # Fit the classifier on the resampled data (if undersampling is used)
        clf.fit(X_train_fold, y_train_fold)
    
        # Predict on the validation fold
        y_pred = clf.predict(X_val_fold)
    
        # Calculate F1 score for the validation fold for class 0
        f1 = f1_score(y_val_fold, y_pred, average='binary', pos_label=0)
        f1_scores.append(f1)
    
        # Compute confusion matrix
        cm = confusion_matrix(y_val_fold, y_pred)
        cm_matrices.append(cm)
    
        # Store the classification report
        report = classification_report(y_val_fold, y_pred, output_dict=True)
        classification_reports.append(report)
        
        # Plot classification results (if needed, replace with your function)
        plot_classification_results(y_val_fold, y_pred)
    
    # Calculate average F1 score across all folds
    avg_f1_score = np.mean(f1_scores)
    print(f"Average F1 score across all folds: {avg_f1_score:.4f}")
```
